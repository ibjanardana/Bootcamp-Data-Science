```python
missing_value = ['na','-']
df = pd.read_csv('austin_weather.csv', na_values=missing_value)
df.info()
```

```python
df = df.sample(1000)
```

```python
# mengubah data float
df = df[['DewPointAvgF','HumidityAvgPercent','WindAvgMPH','TempAvgF']].astype(float)
df.dtypes
```

```python
# tanpa handling missing
plt.figure(figsize=(12,8))
plt.scatter(df['HumidityAvgPercent'], df['DewPointAvgF'], s=df['WindAvgMPH']*20, c=df['TempAvgF'],  cmap='coolwarm')

plt.xlabel('Humidity Average (%)')
plt.ylabel('Dew Point Average (F)')
plt.title('Austin Weather')
plt.show()
```

```python
# dengan handling missing value
df.fillna(method='ffill', inplace=True)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(12,8))
plt.scatter(df['HumidityAvgPercent'], df['DewPointAvgF'], s=df['WindAvgMPH']*20, c=df['TempAvgF'],  cmap='coolwarm')

plt.xlabel('Humidity Average (%)')
plt.ylabel('Dew Point Average (F)')
plt.title('Austin Weather')
plt.show()
```

```python
#Code here
df2['House_Price'].unique()
```

```python
import seaborn as sns
# tanpa handling outliers
sns.boxplot(data=df2, x='House_Price')
```

```python
Q1 = df2['House_Price'].quantile(0.25)
Q3 = df2['House_Price'].quantile(0.75)
IQR = Q3-Q1
lower = Q1 - (1.5*IQR)
upper = Q3 + (1.5*IQR)

print(lower,upper)
```

```python
# with handling outliers
df3 = df2[(df2['House_Price'] < upper) & (df2['House_Price'] > lower)]
df3
```

```python
# boxplot
plt.figure(figsize=(12,8))
sns.boxplot(data=df3, x='House_Price')
```
