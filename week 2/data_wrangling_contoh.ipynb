```python
import pandas as pd

df = pd.read_csv('dataset_18.csv', delimiter=';')
df
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(df):
    # Rename column 'Benda' to 'Benda yang Terbakar'
    df = df.rename(columns={'Benda': 'Benda yang Terbakar'})
    # Rename column 'Nilai' to 'Kerugian (Rupiah)'
    df = df.rename(columns={'Nilai': 'Kerugian (Rupiah)'})
    # Sort by column: 'Kerugian (Rupiah)' (descending)
    df = df.sort_values(['Kerugian (Rupiah)'], ascending=[False])
    #
    df = df.sort_index(axis=1)
    return df

df_clean = clean_data(df.copy())
df_clean.head()
```
