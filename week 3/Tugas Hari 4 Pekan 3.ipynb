```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('vgsales.csv')
df.head()
```

<h2>Soal 1: Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk Global_Sales berdasarkan kategori Genre.

```python
#code here
data_global = df.groupby('Genre')[['Global_Sales']].mean()
data_global
```

<h2>Soal 2: Perbandingan Kuantitatif Pieplot - Part a whole using Pie plot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
#code here

fig,ax = plt.subplots()
ax.pie(data_global['Global_Sales'], labels=data_global.index, autopct='%1.1f%%')

plt.show()
```

<h2>Soal 3: Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
#code here
data_region = df.groupby('Genre')[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].mean()
data_region
```

<h2>Soal 4: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
# code here
import numpy as np

w = 0.13
clr = ['y','grey','violet','b','r','pink','g','y','grey','violet','b','r']
fig, ax = plt.subplots(figsize=(12,8))
x = np.arange(0,8,2)

i = 4
j = 11
for genre in data_region.index: 
    ax.bar(x-w*i, data_region.loc[genre],w, label=genre, color=clr[j])
    i = i-1
    j = j-1

plt.style.use('seaborn')

ax.set_title('Mean Sales Video Games by Genre')
ax.set_ylabel('Mean Sales')
ax.set_xlabel('Region Sales')

ax.set_xticks(x)
ax.set_xticklabels(list(data_region.columns))

fig.legend(loc=1, frameon=True)
plt.show()
```

Insight : Dari bar chart di atas menunjukan bahwa NA_Sales memiliki rata-rata penjualan tertinggi untuk masing-masing genrenya dibandingkan dengan EU_Sales,JP_Sales dan Other_Sales dari setiap masing-masing genrenya.
