```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)

tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig, ax = plt.subplots(figsize=(20,8),nrows=1, ncols=3, sharey=True)
ax[0].plot(x, tan, color='b',label='tanX')
ax[0].set_title('Grafik tanx')
ax[1].plot(x, sin, color='r',label='sinX')
ax[1].set_title('Grafik sinx')
ax[2].plot(x, cos, color='g',label='cosX')
ax[2].set_title('Grafik cosx')
fig.legend(loc='upper center')

plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
#code here
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(x,y, color='b')
ax.set_title('Grafik Cos X')

ax1 = ax.inset_axes([-0.10,0.9,0.28,0.3])
ax1.plot(x,y5, color='y')
ax1.plot(x,y, color='b')
ax1.set_title('Grafik CosX^5')

ax2 = ax.inset_axes([0.8,0.9,0.28,0.3])
ax2.plot(x,y4, color='black')
ax2.plot(x,y, color='b')
ax2.set_title('Grafik CosX^4')

ax3 = ax.inset_axes([0.8,-0.15,0.28,0.3])
ax3.plot(x,y3, color='g')
ax3.plot(x,y, color='b')
ax3.set_title('Grafik CosX^3')

ax4 = ax.inset_axes([-0.1,-0.13,0.28,0.3])
ax4.plot(x,y2, color='r')
ax4.plot(x,y, color='b')
ax4.set_title('Grafik CosX^2')

plt.show()
```
